Random Picture!

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.linspace(-2*np.pi, 2*np.pi, 1000)
X, Y = np.meshgrid(x, y)
Z = (np.sin(X) * 10 + Y**2 / 5)
  
fig, ax = plt.subplots(figsize=(6, 6))
plt.axis('off')
plt.pcolormesh(X, Y, Z, cmap="magma")

for i in range(50):
  x_pt = np.random.randint(-6,6)
  y_pt = np.random.randint(-6,6)
  r = np.random.rand()
  g = np.random.rand()
  b = np.random.rand()
  transparency = np.random.rand()
  colors = (r, g, b)
  for i in range(3):
    radius = np.random.rand()
    circle = plt.Circle((x_pt, y_pt), radius, color=colors, alpha=transparency)
    ax.add_artist(circle)

plt.show()

  
```
