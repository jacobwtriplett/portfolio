```python
# import necessary packages
import matplotlib.pyplot as plt

# define underlying distribution:
#   a six-sided dice representing discrete uniform dist
dice = [1, 2, 3, 4, 5, 6]
results = []

# calculate frequency of outcomes given [four] iid 
#  (independently and identically drawn) samples
i = min(dice)
j = min(dice)
k = min(dice)
l = min(dice)
while i <= len(dice):
  while j <= len(dice):
    while k <= len(dice):
      while l <= len(dice):
        results.append(i+j+k+l)
        l +=1
      k +=1
      l = min(dice)
    j +=1
    k = min(dice)
  i +=1
  j = min(dice)

# write to dictionary and plot
frequency={}
for item in results:
   if item in frequency:
      frequency[item]+=1
   else:
      frequency[item]=1
for key in frequency:
  plt.bar(key, frequency[key])
plt.ylabel('Frequency')
plt.xlabel('Outcome (sum of iid samples)')
plt.show()
```
